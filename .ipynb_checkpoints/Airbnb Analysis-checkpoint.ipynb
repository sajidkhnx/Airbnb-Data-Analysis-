```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('airbnb_ds.csv')
```

```python
df.head(10) 
```

# Check Missing Values

```python
print(df.isnull().sum())
```

```python
df.info()
```

# Handling Missing Values

```python
df['last review'] = pd.to_datetime(df['last review'],errors = 'coerce')
```

```python
df['last review'].info()
```

```python
df.fillna({'reviews per month': 0,'last review': df['last review'].min()}, inplace=True)
```

```python
df.dropna(subset = ['NAME','host name'], inplace= True)
df.isnull().sum()
```

```python
df= df.drop(columns=['house_rules','license'], errors='ignore')
```

```python
df.head()
```

```python
df['price'] = df['price'].astype(str).replace('[\$,]', '', regex=True).str.strip().astype(float)
```

```python
df['service fee'] = df['service fee'].astype(str).replace('[\$,]', '', regex=True).str.strip().astype(float)
```

```python
df.head()
```

# Remove Duplicates

```python
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

# Visualization

```python
# What is ditribution of listing price
```

```python
plt.figure(figsize = (10,6))
sns.histplot(df['price'],bins = 50,kde=True, color='green')
plt.title("Distribution of Listing Price")
plt.xlabel("Price $")
plt.ylabel("Frequency")
plt.show()
```

```python
# What is Relationship between Price and Room Type
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='room type', y='price',hue ='room type',data =df,palette='Set1')
plt.title("Price vs Room Type")
plt.xlabel('Room Type')
plt.ylabel('Price $')
plt.show()
```

```python
#how has number of reviews changes over with time 
```

```python
df['last review'] = pd.to_datetime(df['last review'])
review_over_time = df.groupby(df['last review'].dt.to_period('M')).size()
plt.figure(figsize=(12,6))
review_over_time.plot(kind='line', color='red')
plt.title('Number of reviews over time')
plt.xlabel('Date')
plt.ylabel("Number of review")
plt.show()
```
